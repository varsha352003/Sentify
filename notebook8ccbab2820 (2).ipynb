```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Use a better multi-class sentiment model
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"  # Trained for 1-5 stars
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

def get_star_rating(text):
    """Predicts a rating from 1 to 5 based on sentiment"""
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    probabilities = torch.nn.functional.softmax(logits, dim=-1)

    # Get the star rating (1-based index)
    rating = torch.argmax(probabilities) + 1
    return rating.item()

# Example Reviews
reviews = [
    "I love this movie! It's fantastic.",  
    "The food was terrible, I hated it.",  
    "It was okay, nothing special.",       
    "Absolutely the worst experience ever!",  
    "One of the best things I've watched.",  
    "Meh, not great, not bad.",            
    "Pretty good, I liked it.",            
    "Not bad, but could be better.",       
    "I wouldn’t recommend it, but it’s not the worst.", 
]

# Get ratings
for review in reviews:
    rating = get_star_rating(review)
    print(f"Review: {review}\nPredicted Rating: {rating}/5\n")
```

```python
# Define chunk size
chunk_size = 1000  

# Split data into chunks (ensure copying to avoid SettingWithCopyWarning)
chunks = [sample.iloc[i:i + chunk_size].copy() for i in range(0, sample.shape[0], chunk_size)]

# Process each chunk
for i, chunk in enumerate(chunks):
    print(f"Processing Chunk {i+1}/{len(chunks)}...")

    # Apply sentiment function safely using .loc[]
    chunk.loc[:, "text_score"] = chunk["review/text"].apply(get_star_rating)

    # Save processed chunks incrementally
    if i == 0:
        chunk.to_csv("processed_dataset.csv", mode='w', index=False)  # First write
    else:
        chunk.to_csv("processed_dataset.csv", mode='a', index=False, header=False)  # Append remaining chunks

print("Processing complete! Sentiment scores saved in 'processed_dataset.csv'")
```
